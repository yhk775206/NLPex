```python
# 1. image 
from PIL import Image
img = Image.open('v01.jpg')

# 2. HSV 
import matplotlib
import matplotlib.pyplot as plt
hsv = matplotlib.colors.rgb_to_hsv(img)

h = hsv[:,:,0]
s = hsv[:,:,1]
v = hsv[:,:,2]

# 3. max(v) - min(v)
import numpy as np
print "max(v)-min(v):\t", np.max(v) - np.min(v)

# 4. agerage(v)
print "avg(v):\t", np.average(v)

plt.imshow(img)
plt.show()
```
